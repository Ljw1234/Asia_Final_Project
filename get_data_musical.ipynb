# 콘서트 데이터 크롤링 부분

```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import sqlite3
import re
```

```python
# 순서대로 오리지널/내한공연 ,라이센스, 창작뮤지컬, 넌버별 퍼포먼스
headers = {"user-agent": "Mozilla/5.0"}
consert = ["Ori","Lic","Cre","Non"]
con = ["bal", "Roc"]
url ="http://ticket.interpark.com/TPGoodsList.asp?Ca=Mus&SubCa="
Ori = []
Lic = []
Cre = []
Non = []
for i in consert:
    urls= url+i
    res = requests.get(urls,headers = headers)
    res.raise_for_status()
    soup = BeautifulSoup(res.text,"html.parser")
    data_row = soup.find_all("td",{"class":"RKtxt"})
    for row in data_row:
        data= row.a.text
#         data = [column.get_text().strip() for column in columns]
        print(data)        
        if i == "Ori":
            Ori.append(data)
        elif i == "Lic":
            Lic.append(data)    
        elif i == "Cre":
            Cre.append(data)    
        elif i == "Non":
            Non.append(data)    
  
```

```python
MusicalDf = pd.DataFrame({'Ori': pd.Series(Ori), 'Lic': pd.Series(Lic), 'Cre': pd.Series(Cre), 'Non': pd.Series(Non)})
```
