```python
import requests
from bs4 import BeautifulSoup as bsp
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import sqlite3
import json
from datetime import datetime
```

```python
url = 'https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=656'
reqheaders = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.114 Safari/537.36'}
src=requests.get(url,headers=reqheaders)
cont = src.content.strip()
html = bsp(cont,'html.parser')
tgts = html.select('#Myform')
h1s = tgts[0].select('a.bo3')
h1s[0].text
```

```python
url_1 = h1s[0]['href']
url_1
```

```python
src_1 = requests.get(url_1,headers=reqheaders)
cont_1 = src_1.content.strip()
html_1 = bsp(cont_1,'html.parser')
```

```python
summary_1 = html_1.select('meta')
summary_1[6]['content']
```

```python
wrncom_1 = html_1.select('li.Ere_sub2_title')
```

```python
writier_1 = wrncom_1[0].select('a.Ere_sub2_title')[0].text
writier_1
```

```python
com_1 = wrncom_1[0].select('a.Ere_sub2_title')[1].text
com_1
```

```python
# 경제경영
res = []
url_1s = []
for h1 in h1s:
    r1={}
    tmp_url_1 = h1['href']
    tmp_url_1s = tmp_url_1.split()
#     print(tmp_url_1s)
    url_1s.append(tmp_url_1s)
    for url_1 in url_1s:
        src_1 = requests.get(url_1[0],headers=reqheaders)
        cont_1 = src_1.content.strip()
        html_1 = bsp(cont_1,'html.parser')
        summary_1 = html_1.select('meta')
        wrncom_1 = html_1.select('li.Ere_sub2_title')
        if(summary_1):
            r1['ncontent'] = summary_1[6]['content']
        if(wrncom_1):
            r1['nwriter'] = wrncom_1[0].select('a.Ere_sub2_title')[0].text
            r1['ncompany'] = wrncom_1[0].select('a.Ere_sub2_title')[1].text
    if(h1):
        r1['ntitle'] = h1.text
    if(tmp_url_1):
        r1['nhttp'] = h1['href']
    res.append(list(r1.values()))
        
```

```python
res
```

```python
df1 = pd.DataFrame(res)
```
