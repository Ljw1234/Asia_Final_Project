# 뮤지컬 콘서트 데이터 크롤링 부분

```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import sqlite3
import re
```

```python
# 순서대로 발라드 락 랩 제즈 포크 페스티벌 펜클럽 인디 토크
headers = {"user-agent": "Mozilla/5.0"}
consert = ["Bal","Roc","Rap","Jaz","Por","Fes","Fan","Ind","Tal"]
con = ["bal", "Roc"]
url ="http://ticket.interpark.com/TPGoodsList.asp?Ca=Liv&SubCa="
Bal = []
Roc = []
Rap = []
Jaz = []
Por = []
Fes = []
Fan = []
Ind = []
Tal = []
for i in consert:
    urls= url+i
    res = requests.get(urls,headers = headers)
    res.raise_for_status()
    soup = BeautifulSoup(res.text,"html.parser")
    data_row = soup.find_all("td",{"class":"RKtxt"})
    for row in data_row:
        data= row.a.text
#         data = [column.get_text().strip() for column in columns]
        print(data)        
        if i == "Bal":
            Bal.append(data)
        elif i == "Roc":
            Roc.append(data)    
        elif i == "Rap":
            Rap.append(data)    
        elif i == "Jaz":
            Jaz.append(data)    
        elif i == "Por":
            Por.append(data)    
        elif i == "Fes":
            Fes.append(data)    
        elif i == "Fan":
            Fan.append(data)    
        elif i == "Ind":
            Ind.append(data)
        elif i == "Tal":
            Tal.append(data)
            
            
            
            
            
            
```

```python

ConsertDf = pd.DataFrame({'Bal': pd.Series(Bal), 'Roc': pd.Series(Roc), 'Rap': pd.Series(Rap), 'Jaz': pd.Series(Jaz), 'Por': pd.Series(Por), 'Fes': pd.Series(Fes), 'Fan': pd.Series(Fan), 'Ind': pd.Series(Ind), 'Tal': pd.Series(Tal)})
```
